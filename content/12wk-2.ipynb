### I. 열의 이름 변경

`# 예제`: 열의 이름 출력하고, 열의 이름중 공백(` `)이 있을 경우 언더바(`_`) 로 바꾸자.

```python
df.columns
```

```python
df.columns = [l.replace(' ', '_') for l in df.columns]
```

### J. transform column

`# 예제 `df.Height`열 변환하기 + 변환된 열 할당하기
