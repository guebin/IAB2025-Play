`(2)` 주어진 그림을 올바로 해석한 사람을 모두 골라라.

- 우주: 전체적으로 보면 빨간 점들(수비수)은 오른쪽에, 파란 점들(공격수)은 왼쪽에 몰려 있어. 수비수는 태클 능력이 높은 편이고, 공격수는 낮은 편이라서 포지션별 역할 특성이 분포에 확실히 반영된 것 같아.
- 이애: 확실히 수비수는 대부분 StandingTackle이 50을 넘고, 공격수는 StandingTackle이 50을 넘지 않는 경우가 많네. 
- 다호: 추세선을 살펴보는 것도 흥미로운 일이야. 수비수의 경우, StandingTackle이 높아질수록 시장가치가 급격히 상승해. 아주 뛰어난 태클 능력을 가진 수비수는 시장에서도 그만큼 높은 평가를 받는다는 뜻인 것 같아.
- 주원: 반면에 공격수의 추세선을 보면, 태클 수치가 높아질수록 시장가치는 오히려 낮아지는 경향이 보여. 전체적으로 보면 StandingTackle과 공격수의 시장가치는 반비례 관계처럼 보이더라
