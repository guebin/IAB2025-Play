## 1. mpg자료 

```r
library(tidyverse)
```

```r
df = mpg
```

`-` 칼럼이름들 출력 

```r
names(df)
```

`-` 칼럼설명 

| 칼럼 이름      | 설명                                           | 예시 값            |
|----------------|------------------------------------------------|---------------------|
| manufacturer   | 제조사 이름                                     | audi, ford          |
| model          | 자동차 모델명                                   | a4, f150 pickup 4wd |
| displ          | 배기량 (liters)                                 | 1.8, 5.3            |
| year           | 제조 연도                                       | 1999, 2008          |
| cyl            | 실린더 수                                       | 4, 6, 8             |
| trans          | 변속기 종류 (수동/자동 + 단수)                 | auto(l5), manual(m6)|
| drv            | 구동 방식 (f: 전륜, r: 후륜, 4: 사륜)           | f, r, 4             |
| cty            | 도심 연비 (miles per gallon)                   | 18, 21              |
| hwy            | 고속도로 연비 (miles per gallon)               | 26, 30              |
| fl             | 연료 종류 (p: premium, r: regular, e: ethanol) | p, r, e             |
| class          | 차량 분류 (차량 유형)                          | suv, compact        |

## 2. 시각화 예시 

`#예시1`  

- geom_point: `x=displ`, `y=hwy`

```r
ggplot(df) + geom_point(aes(x=displ, y=hwy))
```

- x가 증가할수록, y는 작아지는 경향이 있음. 
- 배기량이 증가할수록 고속도로 연비가 작아지는 경향이 있음을 관찰

`# 예시10` 

- geom_point: `x=displ`, `y=hwy`, `color=drv` 
- geom_smooth: `x=displ`, `y=hwy`, `color=drv` // `lty=2`, `alpha=0.2`
- geom_smooth: `x=displ`, `y=hwy` // `color="black"`

```r
ggplot(df) + 
geom_point(aes(x=displ, y=hwy, color=drv)) + 
geom_smooth(aes(x=displ, y=hwy,color=drv), alpha=0.2, lty=2) + 
geom_smooth(aes(x=displ, y=hwy),color='black')
```

`# 예시11`

```r
ggplot(df) + 
geom_point(aes(x=displ, y=hwy, color=drv, size=cyl), alpha=0.2) + 
geom_smooth(aes(x=displ, y=hwy,color=drv), alpha=0.2, lty=2) + 
geom_smooth(aes(x=displ, y=hwy),color='black')
```

```r
fig = ggplot(df) 
point = geom_point(aes(x=displ, y=hwy, color=drv, size=cyl), alpha=0.2)
smooth1 = geom_smooth(aes(x=displ, y=hwy,color=drv), alpha=0.2, lty=2) 
smooth2 = geom_smooth(aes(x=displ, y=hwy),color='black')
```
