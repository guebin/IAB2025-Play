```python
df2 = df.drop(["Loaned From", "Marking"],axis=1).dropna()
df2
```

### C. 포지션 정리 

`-` 포지션변환 

```python
df3 = df2.copy()
```

```python
df3['Position'] = [l.split(">")[-1] for l in df2.Position]
df3.Position
```

`-` 포지션치환 

```python
position_dict = {
    'GOALKEEPER':{'GK'},
    'DEFENDER':{'CB','RCB','LCB','RB','LB','RWB','LWB'},
    'MIDFIELDER':{'CM','RCM','LCM','CDM','RDM','LDM','CAM','RAM','LAM','RM','LM'},
    'FORWARD':{'ST','CF','RF','LF','RW','LW','RS','LS'},
    'SUB':{'SUB'},
    'RES':{'RES'}
}
position_dict
```

```python
position = "CAM"
[key for key in position_dict if position in position_dict[key]].pop()
```

```python
df3['Position2'] = [[key for key in position_dict if position in position_dict[key]].pop() for position in df3.Position]
df3.loc[:,['Position','Position2']]
```

### D. 포지션별 능력치 비교 

```python
df3.columns
```

`-` 포지션별로 "Dribbling", "ShotPower", "SlidingTackle" 능력의 평균을 구하라. 

- Dribbling의 평균능력치가 가장 좋은 포지션은 어디인가?
- ShotPower의 평균능력치가 가장 좋은 포지션은 어디인가?
- SlidingTackle의 평균능력치가 가장 좋은 포지션은 어디인가?
- 세 능력치의 평균이 가장 좋은 포지션은?

```python
df3.groupby("Position2")[["Dribbling","ShotPower","SlidingTackle"]].mean()
```

```python
df3.groupby("Position2")[["Dribbling","ShotPower","SlidingTackle"]].mean().mean(axis=1)
```

### F. 주급처리 

```python
df3['Wage'] = [eval(l.replace("€","").replace("K","*1000").replace("M","*1000000")) for l in df2.Wage]
df3
```

### G. 주급분석 

`-` 공격수 포지션의 선수들을 "SlidingTackle"에 따라 4개의 그룹으로 나누어라. "SlidingTackle"능력에 따른 주급차이가 있을까? 

```python
df3.assign(
    StandingTackle_qcut = pd.qcut(df3[df3.Position2 == "FORWARD"].StandingTackle, q=4)
).groupby("StandingTackle_qcut").Wage.mean()
```

`-` 공격수 포지션의 선수들을 슬라이딩 "ShotPower"의 능력에 따라 4개의 그룹으로 나누어라. "ShotPower"능력에 따른 주급차이가 있을까? 

```python
df3.assign(
    ShotPower_qcut = pd.qcut(df3[df3.Position2 == "FORWARD"].ShotPower, q=4)
).groupby("ShotPower_qcut").Wage.mean()
```
