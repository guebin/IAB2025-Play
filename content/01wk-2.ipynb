**`2025수능-6`.** $\cos \left(\frac{\pi}{2} + \theta \right) = -\frac{1}{5}$
일 때, $\frac{\sin\theta}{1 - \cos^2\theta}$
의 값은? **[3점]**

1. $-5$  
2. $-\sqrt{5}$  
3. $0$
4. $\sqrt{5}$
5. $5$

(풀이1)

````r
theta = 2.94
````

````r
cos(pi/2+theta) # 이 값이 -0.2 정도가 되어야 할텐데?
````

반복계산을 하다보니까 $\theta \approx 2.94$ 인듯 

````r
sin(theta)/(1-cos(theta)^2)
````

````r
acos(-1/5) - pi/2
````

````r
sin(0.2013579)/(1-cos(0.2013579)^2)
````

````r
input = -0.2
theta = acos(input) - pi/2
output = sin(theta) / (1-cos(theta)^2)
output
````

이걸 하나로 묶으면.. 

````r
solve <- function(input){
    theta = acos(input) - pi/2
    output = sin(theta)/(1-cos(theta)^2)
    return(output)
}
````

````r
solve(-0.2)
````

## 4. 함수의 선언

`-` 함수의 기본 구조

```r
함수이름 <- function(입력){
    함수의 기능 
    return(출력)
    }
```

````r
solve <- function(input){
    theta = acos(input) - pi/2
    output = sin(theta)/(1-cos(theta)^2)
    return(output)
}
````

````r
solve(-1/5)
````

`-` 아래의 2개의 코드는 `거의` 같다. 

````r
### 1 
solve <- function(input){
    theta = acos(input) - pi/2
    output = sin(theta)/(1-cos(theta)^2)
    return(output)
}
solve(-1/5)

### 2
input = -1/5 
theta = acos(input) - pi/2
output = sin(theta)/(1-cos(theta)^2)
output
````

`-` return을 생략해도 괜찮다. 

````r
solve <- function(input){
    theta = acos(input) - pi/2
    output = sin(theta)/(1-cos(theta)^2)
    return(output)
}
solve(-1/5)
````

````r
solve <- function(input){
    theta = acos(input) - pi/2
    output = sin(theta)/(1-cos(theta)^2)
    output
}
solve(-1/5)
````

````r
solve <- function(input){
    theta = acos(input) - pi/2
    sin(theta)/(1-cos(theta)^2)
}
solve(-1/5)
````
